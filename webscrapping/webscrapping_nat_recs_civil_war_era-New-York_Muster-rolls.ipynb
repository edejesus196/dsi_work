```python
base_url = 'https://www.ancestry.com/search/collections/1965/?birth=1836&count='
```

```python
pages_urls = []
for x in range(50, 116483, 50):
    pages =  base_url + str(x)
    pages_urls.append(pages)
```

```python
pages_urls[0:2]
```

```python
len(pages_urls)
```

```python
results = []

for page in pages_urls:
    r = requests.get(page)
    soup = bs(r.text, 'lxml')
    results.extend(soup.find_all('tr', attrs={'tabindex': '-1'}))

#     name = result.find('span', attrs={'class': 'srchHit'}).text
#     place = result.find_all('td')[4].text
#     url = result.find('a', attrs={'class': 'link icon iconImage'})['href']
#     records.append((name, place, url))
```

```python
results[0:1]
```

```python
len(results)
```

```python
records = []
for result in results:
    name = result.find('td', attrs={'data-label': 'Name'}).text
    place_of_birth = result.find('td', attrs={'data-label': 'Birth Place'}).text
    url = result.find('a', attrs={'class': 'link icon iconImage'})['href']
    records.append((name, place_of_birth, url))
```

```python
columns = ['name', 'place_of_birth', 'link_to_record']
df = pd.DataFrame()

df = pd.DataFrame(records, columns= columns)
```

```python
df.to_csv('ancestry_nat_rec_New-york.csv', index=False, encoding='utf-8')
```
