```python
import requests
import pandas as pd
```

```python
r = requests.get('https://www.ancestry.com/search/collections/2280/?count=50&e-Self-NaturalizationOathofAllegiance=1865&e-Self-NaturalizationOathofAllegiance_x=1-0-0')
```

```python
print(r.text)
```

```python
from bs4 import BeautifulSoup as bs
soup = bs(r.text, 'lxml')
```

```python
results = soup.find_all('tr', attrs={'class': 'record'})
```

```python
type(results)
```

```python
len(results)
```

```python
results[0:1]
```

## Extracting link to record

```python
first_result = results[0]
first_result.find('a', attrs={'class': 'link icon iconImage'})['href']
```

## Extract name

```python
first_result.find('span', attrs={'class': 'srchHit'}).text
```

## Extract Naturalization Place

```python
first_result.find_all('td')[4].text
```

```python
records = []
for result in results:
    name = result.find('span', attrs={'class': 'srchHit'}).text
    place = result.find_all('td')[4].text
    url = result.find('a', attrs={'class': 'link icon iconImage'})['href']
    records.append((name, place, url))
```

```python
records
```

```python
import pandas as pd

columns = ['name', 'place', 'link_to_record']

df = pd.DataFrame(records, columns= columns)
```

```python
df.head()
```

```python
df.to_csv('ancestry_nat_rec_search.csv', index=False, encoding='utf-8')
```
